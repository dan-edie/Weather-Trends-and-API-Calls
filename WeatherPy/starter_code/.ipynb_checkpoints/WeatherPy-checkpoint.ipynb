```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
import sys
sys.path.insert(0, "/Users/Iceda/OneDrive/Desktop/ClassHomework/keys")
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "../output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={weather_api_key}&units={units}&q="
```

```python
# create lists to store data
temp = []
humidity = []
cloudiness = []
wind_speed = []
lat = []
long = []
datetime = []
country = []

print("Beginning Data Retrieval")
print("-----------------------------")

# loops through each city pulled randomly from citipy
counter = 0

for city in cities:
    
    # get the .json() data
    response = requests.get(query_url + city).json()
    
    # append the lists with the data. If no city found, skip
    try:
        temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        country.append(response['sys']['country'])
        datetime.append(response['dt'])
        lat.append(response['coord']['lat'])
        long.append(response['coord']['lon'])
        
        print(f'Processing Record {counter} | {city}')
        
    except:
        print("City not found. Skipping...")
        
    counter += 1

print("-----------------------------")
print("Data Retrieval Complete")
print("-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# there will be more cities than data pulled, so create a dictionary to hold the data pulled
weather_dict = {'City': cities,
                'Country' : country,
                'Datetime' : datetime,
                'Lat' : lat,
                'Lng' : long,
                'Temperature' : temp,
               'Humidity' : humidity,
               'Cloud Coverage' : cloudiness,
               'Wind Speed' : wind_speed}

# now create the dataframe. Any blank data will be NaN
weather_df = pd.DataFrame({ key : pd.Series(value) for key, value in weather_dict.items() })

# drop the cities with Nan values
weather_df = weather_df.dropna()

# set the index name
weather_df.index.name='City_ID'

# convert the Datetime from unix to an actual time
weather_df['Datetime'] = pd.to_datetime(weather_df.Datetime, unit='s')

# output the dataframe into a CSV file
weather_df.to_csv("../output_data/weather_data.csv", index=True)

# show the dataframe
weather_df
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
def regression(x_value, y_value):
    
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_value, y_value)
    regress_values = x_value * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    r_squared = "R-squared = " + str(round(rvalue **2, 2))
    
    plt.figure(figsize = (9, 5))
    fig = plt.scatter(x_value, y_value, facecolors="green", edgecolors="black")
    plt.plot(x_value, regress_values,"r-")
    
    return fig, line_eq, r_squared
```

```python
# Create Northern and Southern Hemisphere DataFrames
north_hemi_df = weather_df.loc[weather_df.Lat >= 0]

south_hemi_df = weather_df.loc[weather_df.Lat < 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
north_max_temp = north_hemi_df.Temperature
north_lat = north_hemi_df.Lat

plot, eq, rsquare = regression(north_lat, north_max_temp)
plt.ylabel("Maximum Temperature ($\degree$ F)")
plt.xlabel("Latitude")
plt.title(f"Norther Hemisphere: Max Temp vs. Latitude Recorded on {date}")
plt.annotate(eq, (0, 10), fontsize=15, color="red")
plt.annotate(rsquare, (0, 0), fontsize=13, color="red")

plt.savefig('../output_data/north_temp_vs_lat.png')
```

* The linear regression provides a clear picture: the further from the equator you move in the northern hemisphere, the less your maximum temperature will be. 
* From the equation of the line, we see that for every degree latitude we move north, the temperature drops by about 1 degree Farenheit.

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
south_max_temp = south_hemi_df.Temperature
south_lat = south_hemi_df.Lat

plot, eq, rsquare = regression(south_lat, south_max_temp)
plt.title(f"Southern Hemisphere: Max Temp vs. Latitude Recored on {date}")
plt.xlabel("Latitude")
plt.ylabel("Maximum Temperature ($\degree$ F)")
plt.annotate(eq, (-15, 43) ,fontsize=15,color="red")
plt.annotate(rsquare, (-13, 37), fontsize=13, color="red")

plt.savefig('../output_data/south_temp_vs_lat.png')
```

* The linear regression for the southern hemisphere also matches what we expect: as we move away from the equator, the temperature will drop. Or, as the plot shows, the closer we get to the equator, the hotter it gets.
* For every degree latitude we move from the south pole, the temperature increases by 0.65 degrees Fahrenheit
